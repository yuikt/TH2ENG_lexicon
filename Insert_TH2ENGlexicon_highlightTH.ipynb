```python
from docx import Document

def read_word_document(filepath):
    """
    Reads a Word document and returns its text content.
    """
    try:
        document = Document(filepath)
        full_text = []
        for paragraph in document.paragraphs:
            full_text.append(paragraph.text)
        return "\n".join(full_text)
    except Exception as e:
        return f"Error reading document: {e}"

# Load Word file
word_filename = 'ฉบับที่ 4 ประกาศ กมช. มาตรฐานและแนวทางส่งเส.docx'
word_path = f'/Users/kornkanoktripanpitak/Documents/Work for Pear/โครงการแปลภาษา/1. เอกสาร Original/{word_filename}'
# document_content = read_word_document(f'/content/{word_filename}') # open with google colab
document_content = read_word_document(word_path)
print(document_content)
```

```python
# Build a Thai→English lexicon and a regex pattern from existing DataFrames df_sheet1..df_sheet6
import re
import pandas as pd
from docx.enum.text import WD_COLOR_INDEX

def build_lexicon(*dfs) -> dict:
    """Combine first two columns (Thai, English) from multiple DataFrames into one dict."""
    lex = {}
    for df in dfs:
        if df is None or df.empty:
            continue
        sub = df.iloc[:, :2].copy()
        sub.columns = ["Thai", "English"]
        sub = sub.dropna(how="any")
        sub["Thai"] = sub["Thai"].astype(str).str.strip()
        sub["English"] = sub["English"].astype(str).str.strip()
        for th, en in sub[["Thai", "English"]].itertuples(index=False):
            if th and en:
                # keep the last seen mapping if duplicates appear across sheets
                lex[" ".join(th.split())] = en  # collapse extra spaces
    return lex

def build_pattern(lexicon: dict) -> re.Pattern:
    """
    Build a regex that matches any Thai term in the lexicon, longest-first,
    and skips cases already followed by parentheses.
    """
    if not lexicon:
        return re.compile(r"(?!x)x")  # never matches
    terms = sorted(lexicon.keys(), key=len, reverse=True)
    escaped = [re.escape(t) for t in terms]
    # Negative lookahead to avoid inserting when parentheses already follow
    return re.compile(r"(" + "|".join(escaped) + r")(?!\s*\()")

def annotate_text(text: str, pattern: re.Pattern, lexicon: dict) -> str:
    return pattern.sub(lambda m: f"{m.group(0)} ({lexicon.get(m.group(0), '')})", text)

def insert_lexicon_into_docx(input_docx: str, output_docx: str, lexicon: dict):
    """
    Open a Word .docx, find exact matches of Thai terms (keys of lexicon),
    and insert their English (values) in parentheses right after each match.
    Saves to a new .docx.
    NOTE: This replaces paragraph/cell text and may remove inline formatting within that run.
    """
    pattern = build_pattern(lexicon)
    doc = Document(input_docx)

    # Paragraphs
    for p in doc.paragraphs:
        new_text = annotate_text(p.text, pattern, lexicon)
        if new_text != p.text:
            p.text = new_text

    # Tables (each cell)
    for tbl in doc.tables:
        for row in tbl.rows:
            for cell in row.cells:
                new_text = annotate_text(cell.text, pattern, lexicon)
                if new_text != cell.text:
                    # Simplest: replace cell text (may drop per-run formatting)
                    cell.text = new_text

    doc.save(output_docx)
    print(f"Saved inserted file to: {output_docx}")


#=============== Below for highlight words ====================#
def copy_format(src_run, dst_run):
    """Best-effort copy of key formatting from src_run to dst_run."""
    dst_run.style = src_run.style
    dst_run.bold = src_run.bold
    dst_run.italic = src_run.italic
    dst_run.underline = src_run.underline
    if src_run.font is not None:
        dst_run.font.name = src_run.font.name
        dst_run.font.size = src_run.font.size
        dst_run.font.color.rgb = getattr(src_run.font.color, "rgb", None)

def annotate_text_to_runs(container, text, base_run, pattern, lexicon):
    """
    Append runs into `container` (Paragraph) that reproduce `text`,
    highlighting matched Thai terms and inserting ' (English)' after them.
    """
    pos = 0
    for m in pattern.finditer(text):
        start, end = m.span()
        thai = m.group(0)
        eng = lexicon.get(thai, "")

        # 1) preceding plain text
        if start > pos:
            r = container.add_run(text[pos:start])
            copy_format(base_run, r)

        # 2) matched Thai term (highlighted)
        r_match = container.add_run(thai)
        copy_format(base_run, r_match)
        r_match.font.highlight_color = WD_COLOR_INDEX.YELLOW

        # 3) inserted English in parentheses (not highlighted)
        if eng:
            r_en = container.add_run(f" ({eng})")
            copy_format(base_run, r_en)

        pos = end

    # 4) trailing text
    if pos < len(text):
        r_tail = container.add_run(text[pos:])
        copy_format(base_run, r_tail)

def process_paragraph(p, pattern, lexicon):
    """Rebuild a paragraph with highlighting + insertions, preserving basic formatting."""
    # Snapshot original runs (text + formatting)
    original_runs = [(r.text, r) for r in p.runs]
    if not original_runs:
        return

    # Clear paragraph
    for r in reversed(p.runs):
        r._element.getparent().remove(r._element)

    # Rebuild
    for text, src_run in original_runs:
        # If no match in this run, keep as-is
        if not pattern.search(text):
            r = p.add_run(text)
            copy_format(src_run, r)
        else:
            annotate_text_to_runs(p, text, src_run, pattern, lexicon)

def process_document(doc: Document, pattern, lexicon):
    # Paragraphs
    for p in doc.paragraphs:
        process_paragraph(p, pattern, lexicon)
    # Tables
    for tbl in doc.tables:
        for row in tbl.rows:
            for cell in row.cells:
                for p in cell.paragraphs:
                    process_paragraph(p, pattern, lexicon)

def insert_lexicon_with_highlight(input_docx: str, output_docx: str, lexicon: dict):
    pattern = build_pattern(lexicon)
    doc = Document(input_docx)
    process_document(doc, pattern, lexicon)
    doc.save(output_docx)
    print(f"Saved inserted & highlighted file to: {output_docx}")
```
